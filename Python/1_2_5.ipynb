```python
#Первый номер,вариант первый(самый эффективный)

#Подключение фрэймворков
import random
import numpy as np

#Тело программы
#Генерация и вывод списка
A = [random.randrange(20) for i in range(10)]
print('Сгенерированный список =',A)

#Нахождение и вывод  повторяющихся значений 

print('Повторяющиеся значения = ',[x for x in A if A.count(x) > 1 ])

#Нахождение и вывод не повторяющихся значений 
print('Не повторяющиеся значения = ',[x for x in A if A.count(x) == 1])

#Вывод средного арефметического значения
print('Среднее арефметическое начальной последовательности = ',np.mean(A))

#Вывод среднего геометрического
SGA = list(A) 
SGAR = pow(np.prod(SGA),(1/len(SGA)))
print('Среднее геометрическое начальной последовательности =',SGAR)

#Генерация и вывод сартированного списка
SA = list(A)
SA.sort()
print('Сортированый  список = ',SA)

```

```python
#Первый номер,вариант второй (без использования простых решений)

#Подключение фрэймворков
import random

#Тело программы
#Генерация и вывод списка
A = list(random.randrange(20) for i in range(10))
print('Сгенерированный список =',A)

#Нахождение и вывод  повторяющихся значений 
PR = []
for x in A: 
  p = 0
  for X in A:
    if  x == X:
      p += 1
  if p > 1:
    PR.append(x)
print('Повторяющиеся значения = ',PR)

#Нахождение и вывод  не повторяющихся значений 
PR = []
for x in A: 
  p = 0
  for X in A:
    if  x == X:
      p += 1
  if p == 1:
    PR.append(x)
print('Не повторяющиеся значения = ',PR)

#Вывод среднего арефметического
p = 0
P = 0
for x in A: 
  for X in A:
    p += 1
    P += X
SAPp = P/p
print('Среднее арефметическое = ',SAPp)

#Вывод среднего геометрического
p = 0
P = 1
for x in A: 
  for X in A:
    p += 1
    P *= X
SGPp = P ** (1/p)
print('Среднее геометрическое = ',SGPp)

#Генерация и вывод сартированного списка

SA = list(A)
SAL =len(SA)
for i in range(SAL - 1):
  for j in range(i+1, SAL):
    if SA[i] > SA[j]:
        SA[i], SA[j] = SA[j], SA[i]
print('Сортированый  список = ',SA)
```
