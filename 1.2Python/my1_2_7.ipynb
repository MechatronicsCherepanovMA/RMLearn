```python

import numpy as np

class RM(object):
  
  def __init__(self):
        self.L1 = 0
        self.L2 = 0
        self.LW1 = 0
        self.HW1 = 0
        self.SW1 = 0
        self.p = 0
        self.t = 0
        self.X = 0
        self.Y = 0
        self.SX = 0
        self.SY = 0
        self.SW1 = 0
        self.dW1 = 0
        self.n = 0 #Почему то не видит данный атрибут в дочернем класе.В моём понимании он должен наследоватся.Подскажите в чем причина.На данный момент про
                    #На данный момент просто определил атребут в дочерних классах.Предпологаю,что не коректно выполняю наследование.
       
  def select(self):
        print("Выберете кинематическую схему манупулятора(1 или 2)")
        self.selectnum = int(input()) 


  def values(self):
        self.L1 = float(input("Введите значение длины для первого звенна в метрах L1(L1 то же  h) = "))
        self.L2 = float(input("Введите значение длины для первого звенна в метрах L2(L2 то же L) = "))
        self.LW1 = float(input("Ввидете меньшее  значение  для диапоззона первого  угла в градусах. LW1 = "))*np.pi/180
        self.HW1 = float(input("Ввидете меньшее  значение  для диапоззона первого  угла в градусах. HW1 = "))*np.pi/180


  def point(self):
        self.p = int(input("Ввидите количество  точек для расчёта в выбранном диапазоне. = "))


  def time(self):
        self.t = float(input("Ввидите период времяни движения манипулятора(c)"))


  def precalc(self):
        self.SW1 = (self.HW1 - self.LW1)/self.t
        self.dW1 = (self.HW1 - self.LW1)/self.p

  def resultprint(self):

        print("Положение конечного звена по оси x =",self.X,)
        print("Положение конечного звена по оси y =", self.Y,)
        print("Угловая скорость конечного звенна по оси х =", self.SX,"рад/с")
        print("Угловая скорость конечного звенна по оси y =", self.SY,"рад/с")

  def calcnum(self):
      self.n += 1
      print("Результат  для ",self.n,"вычесления")
      print("----------------------------------------------------------------------------------")
      

class RM1(RM):

  def __init__(self):
      self.LW2 = 0
      self.HW2 = 0
      self.SW2 = 0
      self.dW2 = 0
      self.n = 0
      
  def values(self):
        RM.values(self)
        self.LW2 = float(input("Ввидете меньшее  значение  для диапоззона второго угла  угла в градусах. LW2 = "))*np.pi/180
        self.HW2 = float(input("Ввидете большее  значение  для диапоззона второго угла  угла в градусах. HW2 = "))*np.pi/180

  def precalc(self):
        RM.precalc(self)
        self.SW2 = (self.HW2 - self.LW2)/self.t
        self.dW2 = (self.HW2- self.LW2)/self.p

  def calc(self):

      for i in range(0,self.p):

        RM.calcnum(self)

        W1 = self.LW1 + self.dW1 * i
        W2 = self.LW2 + self.dW2 * i


        self.X = self.L1 * np.cos(W1) + self.L2 * np.cos(W1 + W2)
        self.Y = self.L1 * np.sin(W1) + self.L2 * np.sin(W1 + W2)
        self.SX = ( - self.L1) * self.SW1 * np.sin(W1) - self.L2 * (self.SW1 + self.SW2) * np.sin(W1 + W2)
        self.SY = self.L1 * self.SW1 * np.cos(W1) + self.L2 * (self.SW1 + self.SW2) * np.cos(W1 + W2)

        RM.resultprint(self)

class RM2(RM):

  def __init__(self):
      self.SL = 0
      self.SH = 0
      self.SS = 0
      self.S = 0
      self.n = 0
  def values(self):
        RM.values(self)
        SL = float(input("Введите нижнее значение диапазона длины подвижного звенна s в метрах = "))
        SH = float(input("Введите верхнее значение диапазона длины подвижного звенна s в метрах = "))

  def precalc(self):

        RM.precalc(self)

        self.SS = (self.SH - self.SL)/self.t
        self.dS = (self.SH - self.SL)/self.p
        
  def calc(self):
        
        for i in range(0,self.p):

          RM.calcnum(self)

          W1 = self.LW1 + self.dW1 * i
          S = self.SH + self.dS * i


          self.X = self.S * np.cos(W1) + self.L2 * np.cos(W1) + self.L1 * np.sin(W1)
          self.Y = self.S * np.sin(W1) + self.L2 * np.sin(W1) - self.L1 * np.cos(W1)
          self.SX = ((- self.S) * np.sin(W1) - self.L2 * np.sin(W1)) + self.L1 * np.cos(W1)* self.SW1 + np.cos(W1) * self.SS
          self.SY = (self.S * np.cos(W1) + self.L2 * np.cos(W1)) + self.L1 * np.sin(W1) * self.SW1 + np.sin(W1) * self.SS 

          RM.resultprint(self)

    
rm = RM()

rm.select()

if rm.selectnum == 1:
   rm = RM1()
   rm.values()
   rm.time()
   rm.point()
   rm.precalc()
   rm.calc()
else:
   rm = RM2()
   rm.values()
   rm.time()
   rm.point()
   rm.precalc()
   rm.calc()
 
```
