```python
#Выполнил Черепанов Михаил Александрович
#При выполнении программы через первую схему запрашивается количество точек для расчёта
#При выполнении программы через вторую схему запрашивается шаг выполнения
#О переменных:Sh - схема W - угл S - скорость L - low H - high Комбинация даст понимание
#Программу переделал.Замечания учёл и исправил.
#Для реалезации использовал numpy as np и его функции для создания масивов из чисел Float.Решение искал в интернете.
#Пример решения пока не просматривал.Возможно после просмотра и проверки программу перепишу.

#Добавление расширений
import math
import numpy as np
# Общий цикл
while True:
#Начало блока для выбора схемы
    print("Выберете какую кинематическую схему вы хотите использвать для расчёта.")
    print("Для расчёта по первой схеме введите 1 по второй схеме ведите 2.")
    print("Для завершения программы введите любое другое значение.")
    scheme = str(input("Ввод: "))
#Конец блока выбора схемы
#Начало блока кода для первой схемы
  
    if scheme == "1":

        print("Вы выбрали кенематическую схему", scheme)

        i1 = str #Переменная для цикл  контрольной проверки введённых данных
# Цикл  контрольной проверки введённых данных
        while i1 != '1':

#Начало блока ввода данных
            
            L1 = float(input("Введите значение длины для первого звенна в метрах L1 = "))

            L2 = float(input("Введите значение длины для первого звенна в метрах L2 = "))

            Sh1LW1 = float (input("Ввидете меньшее  значение  для диапоззона первого  угла в радианах. HL1 = "))

            Sh1HW1 = float(input("Ввидете большее  значение  для диапоззона первого  угла в радианах. HW1 = "))

            Sh1LW2 = float(input("Ввидете меньшее  значение  для диапоззона второго угла  угла в радианах. LW2 = "))

            Sh1HW2 = float(input("Ввидете большее  значение  для диапоззона второго угла  угла в радианах. HW2 = "))

            Sh1SLW1 = float(input("Ввидете меньшее значение диапазона угловой скорости для первого звенав рад/с. SLW1 = "))

            Sh1SHW1 = float(input("Ввидете большее значение диапазона угловой скорости для первого звена рад/с. SHW1 = "))

            Sh1SLW2 = float(input("Ввидете меньшее значение диапазона угловой скорости для второго звена рад/с. SLW2 = "))

            Sh1SHW2 = float(input("Ввидете  большее значение диапазона угловой скорости для второго звена рад/с. SLW2 = "))

            Zn1 = int(input("Ввидите количество  точек для расчёта в выбранном диапазоне. = "))
       
#Конец блока ввода данных

#Начало блока поддтверждения введённой информации.Цикл выше перед блоком ввода данных
            print("Вы ввели следующие значения:")
            print("Длинна первого звенна L1 = ", L1,("м"))
            print("Длинна второго звенна. L2 = ", L2,("м"))
            print("Диапазон для первого угла . W1 = ", Sh1LW1,"-", Sh1HW1,("рад"))
            print("Диапазон для  второго угла . W2 = ", Sh1LW2,"-", Sh1HW2,("рад"))
            print("Диапазон угловой скорости для  первого звена . SW1 = ", Sh1SLW1,"-", Sh1SHW1,("рад/с"))
            print("Диапазон угловой скорости для  второго звена . SW2 = ",  Sh1SLW2,"-", Sh1SHW2,("рад/с"))
            print("количество точек для рассчёта в выбранных диаппахоннах", Zn1)

            print("Что бы произвести вычесление поэтим данным ввидите 1.Что бы записать данные заного введите любое другое")
            i1 = str(input("ввод: "))
       
# Конец блока поддтверждения введённой информации.
#Начало блока подготовки к вычеслению.Запись диапозонов в масив.
        Sh1W1 = list(np.linspace(Sh1LW1, Sh1HW1, num = Zn1))

        Sh1W2 = list(np.linspace(Sh1LW2, Sh1HW2,  num = Zn1))

        Sh1SW1 = list(np.linspace(Sh1SLW1, Sh1SHW1,  num = Zn1))

        Sh1SW2 = list(np.linspace(Sh1SLW2, Sh1SHW2, num = Zn1))

#Конец блока подготовки к вычеслению
#Начало блока вычесления и вывода информации
        r1 = 0 #Переменная для подсчёта циклов расчёта
        for a1,b1,c1,d1 in zip(Sh1W1, Sh1W2, Sh1SW1, Sh1SW2):

                r1 += 1
                print("Расчет для", r1 , "точки последовательности")

                Sh1x = L1*math.cos(a1) + L2 * math.cos(a1+b1)
                print("Положение конччного звена по оси x = ", Sh1x,("градусов"))

                Sh1y = L1 * math.sin(a1) + L2 * math.sin(a1+b1)
                print("Положение конечного звенна по оси y = ", Sh1y,("градусов"))

                Sh1swx = (-L1) * c1 * math.sin(a1) - L2 * (c1 + d1) * math.sin(a1 + b1)
                print("Угловая скорость точки D по оси  x  = ", Sh1swx,("рад/с"))

                Sh1swy = L1 * c1 * math.cos(a1) - L2 * (a1 + d1) * math.cos(a1 + b1)
                print("Угловая скорость точки D по оси у = ", Sh1swy,("рад/с"))
          
#Конец блока вычесления и вывода информации
#Блок кода для второй схемы    
    elif scheme == "2":

        print("Вы выбрали кенематическую схему", scheme)
        
        i2 = str #Переменная для цикл  контрольной проверки введённых данных

        while i2 != '1':
# Цикл  контрольной проверки введённых данных
#Начало блока ввода данных
            

            sl = float(input("Введите нижнее значение диапазона длины первого звенна s в метрах = "))

            sh = float(input("Введите верхнее значение диапазона длины первого звенна s в метрах = "))

            L = float(input("Введите значение длинны второго звена в метрах L = "))

            Sh2LW = float( input("Введите нижнее значение для диапозона  угла первого звенна в радианах W = "))

            Sh2HW = float(input("Введите верхнее значение для диапозона  угла первого звенна в радианах W = "))
            
            Sh2SLW = float(input("Введите нижнее значение диапозона для  угловой скорости первого звенна в рад/с SW = "))

            Sh2SHW = float(input("Введите верхнее значение диапозона для  угловой скорости первого звенна в в рад/с SW = "))

            h = float(input("Введите значение длины h в метрах = "))

            ssl = float(input("Введите нижнее значение диапозона  скороссти движения второго звена в м/с ss = "))

            ssh = float(input("Введите верхнее значение диапозона  скороссти движения второго звена в м/с ss = "))

            Zn2 = float(input("Введите шаг изменения значений диапозонов для введёных данных = "))

#Конец блока ввода данных
#Начало блока поддтверждения введённой информации.Цикл выше перед блоком ввода данных
            print("Вы ввели следующие значения:")
            print("Диапозон значений длины для звена s = ", sl,"-",sh,("м"))
            print("Длинна первого звенна L = ", L ,("м"))
            print("Диапозон  значений для  угла W = ", Sh2LW," - ",Sh2SHW,("рад"))
            print("Диапозон  значений для угловой скорости первого звенна SW = ", Sh2SLW," - ",Sh2SHW,("рад/с"))
            print("Длинна  звенна h = ", h,("м"))
            print("Диапозон  значений для cкорости перемещения звенна s = ", ssl,"-",ssh,("рад/с"))
            print("Шаг изменения значений диапозонов для введёных данных =", Zn2)

            print("Что бы произвести вычесление поэтим данным ввидите 1.Что бы записать данные заного введите любое другое")
            i2 = str(input("ввод: "))

#Конец блока поддтверждения введённой информации.
#Начало блока подготовки к вычеслению
        Sh2S = list(np.arange(sl,sh,Zn2))

        Sh2SS = list(np.arange(ssl, ssh,Zn2))

        Sh2W = list(np.arange(Sh2LW,Sh2HW,Zn2))
        
        Sh2SW = list(np.arange(Sh2SLW,Sh2SHW,Zn2))
# Конец блока подготовки к вычеслению
# Начало блока вычесления и вывода информации
        R1 = 0
        for A1, B1, C1, D1 in zip(Sh2S, Sh2SS, Sh2W, Sh2SW):

            R1 += 1
            print("Расчет для", R1, "шага последовательности")
            Sh2x = A1 * math.cos(C1) + L * math.cos(C1) + h * math.sin(C1)
            print("Положение конечного звена по оси x =", Sh2x,"градусов")

            # Sh2 вычесление (X)
            Sh2y = A1 * math.sin(C1) + L * math.sin(C1) - h * math.cos(C1)
            print("Положение конечного звена по оси y =", Sh2y,"градусов")

            # Sh2 вычесление (swx)
            Sh2swx = ((-A1) * math.sin(C1) - L * math.sin(C1) + h *math.cos(C1)) * D1 + math.cos(C1) * B1
            print("Угловая скорость конечного звенна по оси y =", Sh2swx,"рад/с")

            # Sh2 вычесление (swx)
            Sh2swy = (A1 * D1 * math.cos(C1) - L * math.sin(C1) + h *math.cos(C1)) * D1 + math.sin(C1) * B1
            print("Угловая скорость конечного звенна по оси y =", Sh2swy,"рад/с")
# Конец блока вычесления и вывода информации

    else:

        print("Вы остановили выполнение программы")
        break
```
