```python
#   
#          
#        
# :Sh -  W -  S -  L - low H - high   
# .   .
#   numpy as np         Float.   .
#    .      .

# 
import math
import numpy as np
#  
while True:
#    
    print("        .")
    print("      1     2.")
    print("      .")
    scheme = str(input(": "))
#   
#     
  
    if scheme == "1":

        print("   ", scheme)

        i1 = str #       
#      
        while i1 != '1':

#   
            
            L1 = float(input("        L1 = "))

            L2 = float(input("        L2 = "))

            Sh1LW1 = float (input("           . HL1 = "))

            Sh1HW1 = float(input("           . HW1 = "))

            Sh1LW2 = float(input("            . LW2 = "))

            Sh1HW2 = float(input("            . HW2 = "))

            Sh1SLW1 = float(input("         /. SLW1 = "))

            Sh1SHW1 = float(input("         /. SHW1 = "))

            Sh1SLW2 = float(input("         /. SLW2 = "))

            Sh1SHW2 = float(input("          /. SLW2 = "))

            Zn1 = int(input("        . = "))
       
#   

#    .     
            print("   :")
            print("   L1 = ", L1,(""))
            print("  . L2 = ", L2,(""))
            print("    . W1 = ", Sh1LW1,"-", Sh1HW1,(""))
            print("     . W2 = ", Sh1LW2,"-", Sh1HW2,(""))
            print("       . SW1 = ", Sh1SLW1,"-", Sh1SHW1,("/"))
            print("       . SW2 = ",  Sh1SLW2,"-", Sh1SHW2,("/"))
            print("      ", Zn1)

            print("       1.       ")
            i1 = str(input(": "))
       
#     .
#    .   .
        Sh1W1 = list(np.linspace(Sh1LW1, Sh1HW1, num = Zn1))

        Sh1W2 = list(np.linspace(Sh1LW2, Sh1HW2,  num = Zn1))

        Sh1SW1 = list(np.linspace(Sh1SLW1, Sh1SHW1,  num = Zn1))

        Sh1SW2 = list(np.linspace(Sh1SLW2, Sh1SHW2, num = Zn1))

#    
#     
        r1 = 0 #    
        for a1,b1,c1,d1 in zip(Sh1W1, Sh1W2, Sh1SW1, Sh1SW2):

                r1 += 1
                print(" ", r1 , " ")

                Sh1x = L1*math.cos(a1) + L2 * math.cos(a1+b1)
                print("     x = ", Sh1x,(""))

                Sh1y = L1 * math.sin(a1) + L2 * math.sin(a1+b1)
                print("     y = ", Sh1y,(""))

                Sh1swx = (-L1) * c1 * math.sin(a1) - L2 * (c1 + d1) * math.sin(a1 + b1)
                print("   D    x  = ", Sh1swx,("/"))

                Sh1swy = L1 * c1 * math.cos(a1) - L2 * (a1 + d1) * math.cos(a1 + b1)
                print("   D    = ", Sh1swy,("/"))
          
#     
#        
    elif scheme == "2":

        print("   ", scheme)
        
        i2 = str #       

        while i2 != '1':
#      
#   
            

            sl = float(input("       s   = "))

            sh = float(input("       s   = "))

            L = float(input("       L = "))

            Sh2LW = float( input("           W = "))

            Sh2HW = float(input("           W = "))
            
            Sh2SLW = float(input("           / SW = "))

            Sh2SHW = float(input("            / SW = "))

            h = float(input("   h   = "))

            ssl = float(input("          / ss = "))

            ssh = float(input("          / ss = "))

            Zn2 = float(input("        = "))

#   
#    .     
            print("   :")
            print("     s = ", sl,"-",sh,(""))
            print("   L = ", L ,(""))
            print("      W = ", Sh2LW," - ",Sh2SHW,(""))
            print("        SW = ", Sh2SLW," - ",Sh2SHW,("/"))
            print("   h = ", h,(""))
            print("    c   s = ", ssl,"-",ssh,("/"))
            print("       =", Zn2)

            print("       1.       ")
            i2 = str(input(": "))

#    .
#    
        Sh2S = list(np.arange(sl,sh,Zn2))

        Sh2SS = list(np.arange(ssl, ssh,Zn2))

        Sh2W = list(np.arange(Sh2LW,Sh2HW,Zn2))
        
        Sh2SW = list(np.arange(Sh2SLW,Sh2SHW,Zn2))
#     
#      
        R1 = 0
        for A1, B1, C1, D1 in zip(Sh2S, Sh2SS, Sh2W, Sh2SW):

            R1 += 1
            print(" ", R1, " ")
            Sh2x = A1 * math.cos(C1) + L * math.cos(C1) + h * math.sin(C1)
            print("     x =", Sh2x,"")

            # Sh2  (X)
            Sh2y = A1 * math.sin(C1) + L * math.sin(C1) - h * math.cos(C1)
            print("     y =", Sh2y,"")

            # Sh2  (swx)
            Sh2swx = ((-A1) * math.sin(C1) - L * math.sin(C1) + h *math.cos(C1)) * D1 + math.cos(C1) * B1
            print("      y =", Sh2swx,"/")

            # Sh2  (swx)
            Sh2swy = (A1 * D1 * math.cos(C1) - L * math.sin(C1) + h *math.cos(C1)) * D1 + math.sin(C1) * B1
            print("      y =", Sh2swy,"/")
#      

    else:

        print("   ")
        break
```
